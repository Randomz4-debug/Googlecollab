# ***Calculus***

```python
import matplotlib.pyplot as plt
import math

limit = 50

#For single point if y = x^n
def cal_curve_slope_one_point(limit, pow):
  y_values = []
  x_values = []
  slope_val = []
  for i in range(1, limit):
    x = i
    y = math.pow(x, pow)      # y = x^2
    slope = pow*x             # slope = 2x
    y_values.append(y)
    x_values.append(x)
    slope_val.append(slope)
  return [x_values, y_values, slope_val]

#For 2 values if y = x^n
def cal_curve_slope_two_point(limit, height, pow):
  x1_val = []
  y1_val = []
  x2_val = []
  y2_val = []
  slope_val = []
  for i in range(1, limit):
    x1 = i
    x2 = x1 + height
    y1 = math.pow(x1, pow)
    y2 = math.pow(x2, pow)
    x1_val.append(x1)
    x2_val.append(x2)
    y1_val.append(y1)
    y2_val.append(y2)

    slope = (y2 - y1) / (x2 - x1)
    slope_val.append(slope)
  return [x1_val, y1_val, x2_val, y2_val, slope_val]

def plot(x, y, x_label, y_label, title, grid):
  plt.plot(x, y)
  plt.xlabel(x_label)
  plt.ylabel(y_label)
  plt.title(title)
  plt.grid(grid)
  plt.show()
```

```python
val = cal_curve_slope_two_point(limit, height=0.1, pow=5)
plot(val[0], val[1], 'x2', 'y2', 'None', True)
plot(val[2], val[3], 'x2', 'y2', 'None', True)
print([int(s) for s in val[4]])
```
